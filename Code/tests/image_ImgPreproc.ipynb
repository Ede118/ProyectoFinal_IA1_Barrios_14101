```python
from pathlib import Path
import cv2 as cv
import matplotlib.pyplot as plt

from Code.types import PROJECT_ROOT
from Code.image.ImgPreproc import ImgPreproc, ImgPreprocCfg

# 1) Eleg la imagen
img_path = PROJECT_ROOT / "DataBase" / "data" / "images1" / "Tornillo" / "Tornillo001.jpg"
img_bgr = cv.imread(str(img_path))
assert img_bgr is not None, f"No pude leer {img_path}"

# 2) Ejecut el nuevo preprocesamiento
cfg = ImgPreprocCfg()                     # cambi parmetros si quers
pre = ImgPreproc(cfg)
out = pre.process(img_bgr)                # devuelve PreprocOutput

# 3) Arm una visualizacin rpida
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].set_title("Original")
axs[0, 0].imshow(cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB))
axs[0, 0].axis("off")

axs[0, 1].set_title("Mscara")
axs[0, 1].imshow(out.mask, cmap="gray")
axs[0, 1].axis("off")

overlay = img_bgr.copy()
if out.meta is not None:
    cv.drawContours(overlay, [out.meta.contour], -1, (0, 255, 0), 2)
    box = cv.boxPoints(out.meta.rect).astype(int)
    cv.polylines(overlay, [box], True, (255, 0, 0), 2)

axs[1, 0].set_title("Overlay")
axs[1, 0].imshow(cv.cvtColor(overlay, cv.COLOR_BGR2RGB))
axs[1, 0].axis("off")

axs[1, 1].set_title("Crop alineado")
axs[1, 1].imshow(out.img, cmap="gray")
axs[1, 1].axis("off")

plt.tight_layout()
plt.show()
```
